```python
movies = movies.rename(columns={'title_x':'title','cast_y':'cast','crew_y':'crew'})
```

```python
## keep the requirement column and drop others
# keep - genres,id,keywords,overview,cast,crew,title
df = movies[['id','title','overview','genres','keywords','cast','crew']]
```

```python
df.head(1)
```

```python
import ast
## evaluates a string into a list
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
## find the missing value
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
df.head(1)
```

```python
df['keywords'] = df['keywords'].apply(convert)
df.head(1)
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
```

```python
df['cast'] = df['cast'].apply(convert3)
```

```python
df['cast'].head()
```

```python
df['cast'] = df['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
df['crew'] = df['crew'].apply(fetch_director)
```

```python
df.head(1)
```

```python
df['overview'].isna().sum()
```

```python
df['overview'] = df['overview'].apply(lambda x : x.split())
```

```python
df.head(1)
```

```python
## remove the spaces betweeen words
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(" ","") for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df.head()
```

```python
df['tags'] = df['overview'] + df['genres'] + df['keywords'] + df['cast'] + df['crew']
```

```python
df.head()
```

```python
new_df = df[['id','title','tags']]
```

```python
new_df.head()
```

```python
## convert list to string
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
%pip install nltk
import nltk
```

```python
##['love','loved','loving'] -> ['love','love','love']
# convert same type word into a single word

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)     
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
%pip install scikit-learn
from sklearn.feature_extraction.text import CountVectorizer
```

```python
## extract most frequient word on 5000 words
cv = CountVectorizer(max_features=5000,stop_words="english")
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
## calculate cosine similarity

from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
 ## user give a move name then recomend 5 similar movie
 def recomend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distance = similarity[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recomend('Avatar')
```
